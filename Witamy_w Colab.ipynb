```
import tensorflow as tf
import numpy as np
```

```
def f(x):
  return 12*x+3
x = np.arange(-3,3,0.1)
y=np.array([f(_) for _ in x])
print(y)
```

```
input_layer = tf.keras.Input(shape=(1,))
model=tf.keras.models.Sequential([input_layer, tf.keras.layers.Dense(1)])
model.summary()
```

```
model.compile(loss='mse',optimizer=tf.keras.optimizers.Adam(learning_rate=0.1))
```

```
model.fit(x,y,epochs=100)
```

```
predict=model.predict(x)
print(y)
print(predict.flatten())
```

```
x_1=np.arange(-300,300,1)
print(x_1)
y_1=np.array([f(_) for _ in x_1])
print(y_1)
```

```
predict_1=model.predict(x_1)
print(y_1)
print(predict_1.flatten())
```

```
diff=y_1-predict_1.flatten()
rounded_diff=np.round(diff,5)
format_diff=[f'{val:.2f}' for val in rounded_diff]
matrix_diff=np.array(format_diff).reshape(-1,10)
print(matrix_diff)
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(x_1,y_1,label='y_1 True Value', color='blue')
plt.xlabel('x_1')
plt.ylabel('y_1')
plt.title('True Value')
plt.legend()
plt.grid(True)
plt.show
```

```
plt.figure(figsize=(8,6))
plt.plot(x_1,predict_1,label='predict_1 Predicted Value', color='green')
plt.xlabel('x_1')
plt.ylabel('y_1')
plt.title('Prediction Value')
plt.legend()
plt.grid(True)
plt.show
```
